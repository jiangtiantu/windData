```python
# coding: utf-8

from WindPy import w

w.start()

#获取A股代码和简称
def getStockList(tr_date):
    #return datastruct [[datetime1,datetime2,....],[stockcode1,stockcode2,...],[stockname1,stockname2,...]]
    stockdata = w.wset('sectorconstituent',"date=" + tr_date + ";sectorid=a001010100000000")
    return stockdata.Data

#获取A股一览表
def getListedSecurityGeneralView():
    # return datastruct [[sec_code,...],[sec_name,...],[close,...],[total_market,...],[mkt_cap_flo,...],
    #[trade_status,...],[last_trade,...],[ipo_date,...],[province,...],[sec_type,...],[list_boad,...],[exchange]]
    return w.wset("listedsecuritygeneralview","sectorid=a001010100000000").Data

#获取全部A股交易状态，涨跌停状态，涨跌幅
def getSecStatus(tr_date):
    sec_code_list = getStockList(tr_date)[1]
    return w.wss(sec_code_list,"trade_status,maxupordown,pct_chg","tradeDate=" + tr_date+";cycle=D").Data

#返回指数涨跌幅
def getIndexChg(tr_date):
    #return [[上证综指，深圳成指，中小板，创业板，沪深300，中证500，上证50]]
    return w.wss("000001.SH,399001.SZ,399005.SZ,399006.SZ,000300.SH,000905.SH,000016.SH", "pct_chg","tradeDate="+tr_date+";cycle=D").Data[0]

#返回沪市成交额和深市成交额
def getAmt(tr_date):
    return w.wss("000001.SH,399106.SZ", "amt","tradeDate="+tr_date+";cycle=D").Data[0]

#获取停牌个数，交易个数，涨停个数，跌停个数，上涨个数，下跌个数，涨幅大于5%的个数，跌幅大于5%的个数,指数涨跌幅，两市成交量
def getTraStatus(tr_date):
    item ={}
    status = getSecStatus(tr_date)
    item["halt_stocks"] = status[0].count(u"停牌一天")
    item["trade_stocks"] = status[0].count(u"交易")
    item["raise_limits"] = status[1].count(1) 
    item["down_limits"] = status[1].count(-1)
    item["raise_stocks"]  = [x for x in status[2] if x>0 and isinstance(x,float)].__len__()
    item["down_stocks"] = [x for x in status[2] if x<0 and isinstance(x,float)].__len__()   
    item["raise5_stocks"]  = [x for x in status[2] if x>5 and isinstance(x,float)].__len__()
    item["down5_stocks"] = [x for x in status[2] if x<-5 and isinstance(x,float)].__len__()
    
    indexchg = getIndexChg(tr_date)
    item["shangzhengzongzhi"] = round(indexchg[0],2)
    item["shenzhenchengzhi"] = round(indexchg[1],2)
    item["zhongxiaobanzhi"] = round(indexchg[2],2)
    item["chuangyebanzhi"] = round(indexchg[3],2)
    item["hs300"] = round(indexchg[4],2)
    item["zz500"] = round(indexchg[5],2)
    item["sz50"] = round(indexchg[6],2)
    
    amt = getAmt(tr_date)
    item["hushi"] = round(amt[0]/100000000,2)
    item["shenshi"] = round(amt[1]/100000000,2)
    
    return item

def getData(tr_date):
    a = getTraStatus(tr_date)
    p1 = '今日涨停{0}只，跌停{1}只；涨幅超过5%的{2}只，跌幅超过5%的{3}只；上涨{4}只，下跌{5}只。'.format(a["raise_limits"],a["down_limits"],a["raise5_stocks"],a["down5_stocks"],a["raise_stocks"],a["down_stocks"])
    p2 = '指数方面：上证综指（{0}%），深成指（{1}%），创业板（{2}%）'.format(a["shangzhengzongzhi"],a["shenzhenchengzhi"],a["chuangyebanzhi"])
    p3 = '成交量方面：沪市成交{0}亿，深市成交{1}亿，两市共成交{2}亿'.format(a["hushi"],a["shenshi"],a["hushi"]+a["shenshi"])
    print "大盘概览"
    print p1
    print p2
    print p3
    return [p1,p2,p3]

if __name__=="__main__":
    date = "20170614"
    print date
    rlist = getData(date)
    f = file("result.txt","w+")
    f.writelines(rlist)
    f.close()
    w.close()
```
