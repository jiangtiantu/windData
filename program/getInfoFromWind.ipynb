```python
#coding:urf-8
import pandas as pd
import datetime

from WindPy import w
w.start()

#根据日期和板块ID获取板块成分股信息
"""
参数：交易日期 & 板块ID
返回值：wind_code、sec_name
返回格式：pandas.DataFrame
    td = "20170522"
    sectorid = "a001010100000000"
    df = getBKChenFen(td,sectorid)
"""
def getBKChengFen(td,sectorid):
    wsd_data = w.wset("sectorconstituent","date="+td+";sectorid=" +sectorid)
    df = pd.DataFrame(wsd_data.Data,wsd_data.Fields)
    return df.T

"""
参数：td、windcode
返回值：wind_code、sec_name
返回格式：pandas.DataFrame
    td = "20170522"
    windcode = "000300.SH"
    df = getIndexChenFen(td,windcode)
"""
def getIndexChenFen(td,windcode):
    wsd_data = w.wset("sectorconstituent","date="+td+";windcode=" +windcode)
    df = pd.DataFrame(wsd_data.Data,wsd_data.Fields)
    return df.T

"""
参数：td、windcode
返回值：wind_code、sec_name、i_weight
返回格式：pandas.DataFrame
    td = "20170522"
    windcode = "000300.SH"
    df = getIndexWeight(td,windcode)
"""
def getIndexWeight(td,windcode):
    wsd_data = w.wset("indexconstituent","date="+td+";windcode=" +windcode)
    df = pd.DataFrame(wsd_data.Data,wsd_data.Fields)
    return df.T

#根据日期获取停复牌信息
"""
参数：st_date、end_date、flag
df = getTFP(st_date,end_date,"停牌")
"""
def getTFP(st_date,end_date,flag):
    if flag=="停牌":
        codeid = "tradesuspend"
    elif flag=="复牌":
        codeid = "traderesume"        
    wsd_data = w.wset(codeid,"startdate="+st_date+";enddate="+end_date)
    df = pd.DataFrame(wsd_data.Data,wsd_data.Fields)
    return df.T

#获取融资融券信息
"""
w.wset("margintradingunderlying","date=2017-05-22")

w.wset("shortsellingunderlying","date=2017-05-22")
"""
def getRZRQ(flag,td):
    if flag=="融资":
        codeid = "margintradingunderlying"
    elif flag == "融券":
        codeid="shortsellingunderlying"
    wsd_data = w.wset(codeid,"date="+td)
    df = pd.DataFrame(wsd_data.Data,wsd_data.Fields)
    return df.T

if __name__=="__main__":
    td = "20170522"
    sectorid = "a001010100000000"
    windcode = "000300.SH"
    st_date = "20170203"
    end_date = "20170522"
    flag = "复牌"
#    df = getBKChenFen(td,sectorid)
#    df = getIndexWeight(td,windcode)
#    df = getTFP(st_date,end_date,flag)
    df = getRZRQ("融资",td)
    print df
```

```python
#coding:urf-8
from WindPy import w
import pandas as pd
import datetime

w.start()

def getBKChenFen(td,sectorid):
    wsd_data = w.wset("sectorconstituent","date="+td+";sectorid=" +sectorid)
    df = pd.DataFrame(wsd_data.Data,wsd_data.Fields)
    return df.T

def getHQ(tickers,fields,td,cycle):
    if type(tickers)=='list':
        tickers = ','.join(tickers)
    if type(fields)=="list":
        fields = ','.join(fields)
        
    wss_data = w.wss(tickers,fields,"tradeDate="+td+";cycle="+cycle)
    wss_data
    df = pd.DataFrame(wss_data.Data,wss_data.Fields)
    return df.T


#----------------------------------------------------统计行业涨跌幅----------------------------------------
if __name__=="__main__":
    td='20170801'
    sw_1 = "a39901011g000000"
    sw_2 = "a39901011h000000"
    zx_3 = "a39901012g000000"
    
    code_sw_1 = list(getBKChenFen(td,sw_1).wind_code)        #获取申万一级行业代码
#    code_sw_2 = list(getBKChenFen(td,sw_2).wind_code)        #获取申万二级级行业代码
#    code_zx_3 = list(getBKChenFen(td,zx_3).wind_code)        #获取中信三级级行业代码
    
    fields = ['sec_name','pct_chg']
    cycle = 'D'

    df_sw1 = getHQ(code_sw_1,fields,td,cycle).sort("PCT_CHG",ascending=False)
#    df_sw2 = getHQ(code_sw_2,fields,td,cycle).sort("PCT_CHG",ascending=False)
#    df_zx3 = getHQ(code_zx_3,fields,td,cycle).sort("PCT_CHG",ascending=False)

    print df_sw1.head(10)
#    print df_sw2.head(10)
#    print df_zx3.head(10)
```
