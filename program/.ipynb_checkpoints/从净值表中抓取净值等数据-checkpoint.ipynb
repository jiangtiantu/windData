### 从宏观对冲产品净值表中获取净值数据

```python
# coding: utf-8

import os
import xlrd

def getFiles(path):
    flist = []
    for rt,dirs,files in os.walk(fpath):
        for f in files:
 #           print f.decode('gbk',"ignore")
            flist.append(rt+'\\'+f)
    return flist

def getDataFexcel(filename):
    res = {}
    data = xlrd.open_workbook(filename)
    table = data.sheets()[0]
    nrow = table.nrows
    res['tdate'] = table.cell(2,0).value[-10:]
    res['jz'] = table.cell(2,7).value[-6:]
    res['cyb']=0
    res['yhrl']=0
    res['sh']=0
    res['sz']=0
    res['gpzcw']=0
    res['cyb']=0
    res['yhrl']=0
    for i in  range(nrow):    
#        print len(table.cell(i,1).value),table.cell(i,1).value
        if table.cell(i,1).value == u'上交所A股':
            res['sh'] = table.cell(i,8).value
        if table.cell(i,1).value== u'深交所A股':
            res['sz'] = table.cell(i,8).value
        if table.cell(i,1).value == u'股票投资':
            res['gpzcw'] = table.cell(i,8).value
        if table.cell(i,1).value==u'创业板':
            res['cyb'] = table.cell(i,8).value
        if table.cell(i,1).value == u'银华日利':
            res['yhrl'] = table.cell(i,8).value
     
    return res


fpath = 'C:\jf\hg'   
ls=getFiles(fpath)

file_obj=open('result.txt','w+')
print "日期 净值 沪市仓位 深市仓位 创业板基金 货币基金 股票总仓位   "
for l in ls:
    da = getDataFexcel(l)
    print da['tdate'],da['jz'],da['sh'],da['sz'],da['cyb'],da['yhrl'],da['gpzcw']
    file_obj.write(str(da['tdate'])+'\t'+str(da['jz'])+'\t'+str(da['sh'])+'\t'+str(da['sz'])+str(da['cyb'])+'\t'+str(da['yhrl'])+'\t'+'\t'+str(da['gpzcw'])+'\t'+'\n')

file_obj.close()
```

### 从择时对冲净值表中获取净值数据

```python
#coding=utf-8

import xlrd
import os

def getFileList(filepath):
	filelist = os.listdir(filepath)
	return filelist
	pass
def getData(filename):
	item={}
	item['date']=""
	item['流通股票投资合计']=0
	item['股指期货投资合计']=0
	item['净值']=0
	table=xlrd.open_workbook(filename).sheets()[0]

	for i in range(table.nrows):
		if table.row(i)[0].value == u'流通股票投资合计':
			item['date']=table.row(2)[0].value
			item['流通股票投资合计']=table.row(i)[8].value
		if table.row(i)[0].value==u'股指期货投资合计：':
			item['股指期货投资合计']=table.row(i)[8].value
		if table.row(i)[0].value == u'今日单位净值：':
			item['净值'] = table.row(i)[1].value
			print item['净值']
        
	return item


filepath=r'C:\jf\hg'
items=[]
for f in getFileList(filepath):
	print '------------'+str(f)
	fn=filepath+'\\'+f
	print fn
	items.append(getData(fn))
	print '-=-=-=-=-=-==-='
print len(items)

file_obj=open('result.txt','w+')

for it in items:
	date1=it['date'][5:]
	stockvol='股票仓位:'.decode('utf-8').encode('gbk')+str(it['流通股票投资合计']*100)+'%'
	fuvol='期货仓位:'.decode('utf-8').encode('gbk')+str(it['股指期货投资合计']*100)+'%'
	ck='敞口：'.decode('utf-8').encode('gbk')+str(it['流通股票投资合计']*100+it['股指期货投资合计']*100)+'%'
	jz = str(it['净值'])

	date1=date1.encode('utf-8')
	stockvol=stockvol.decode('gbk').encode('utf-8')
	fuvol=fuvol.decode('gbk').encode('utf-8')
	ck=ck.decode('gbk').encode('utf-8')
	# print it['流通股票投资合计']
	# print it['股指期货投资合计']
	# print '股票仓位:'.decode('utf-8').encode('gbk')+str(it['流通股票投资合计']*100)+'%'
	# print '期货仓位:'.decode('utf-8').encode('gbk')+str(it['股指期货投资合计']*100)+'%'
	# print '敞口：'.decode('utf-8').encode('gbk')+str(it['流通股票投资合计']*100+it['股指期货投资合计']*100)+'%'
	file_obj.write(date1+'\t'+stockvol+'\t'+fuvol+'\t'+ck+'\t'+jz+'\n')
print '===ok==='

file_obj.close()
```
